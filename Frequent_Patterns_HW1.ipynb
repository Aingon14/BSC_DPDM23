```python
# แปลงคอลัมน์ให้เป็น datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
#เปลี่ยนรูปแบบให้เป็น เดือน / วัน / ปี
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%m/%d/%y')
```

```python
# สร้างคอลัมน์ 'Date' โดยดึงข้อมูลเฉพาะวันที่
data['Date'] = data['InvoiceDate'].dt.date
data
```

```python
data['month'] = data['InvoiceDate'].dt.month
```

```python
data['day'] = data['InvoiceDate'].dt.day
```

# ข้อ 1 หาข้อมูลของช่วงเวลา Chrhistmas & New Year (1 Dec - 31 Dec)

```python
# ดูช่วงเวลาคริสต์มาสกับปีใหม่
Christm_Newy = data[data['month']==12]
Christm_Newy
```

```python
# แปลง data  ให้อยู่ในรูปแบบ Transaction data

trans_Christm_Newy = Christm_Newy.groupby('InvoiceNo')['StockCode'].apply(list)
```

```python
# หาความถี่ของสินค้าที่ถูกซื้อซ้ำๆกันในช่วงเวลาคริสต์มาสกับปีใหม่ โดยมีค่า min support คือ 8%
frequentItemsets= list(apriori(trans_Christm_Newy,min_support = 0.08))
frequentItemsets
```

# ข้อ 2 หาข้อมูลของประเทศในทวีป Europe

```python
#แยกประเทศที่อยู่ในทวีปยุโรป

Europe = [
    'Austria', 'Belgium', 'Channel Islands','Cyprus', 'Czech Republic', 'Denmark', 'EIRE',
    'European Community', 'Finland', 'France', 'Germany', 'Greece', 'Iceland', 'Italy',
    'Lithuania', 'Malta', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Spain', 'Sweden',
    'Switzerland', 'United Kingdom'
]
```

```python
EU_data = data[data['Country']==""]
for i in range(len(Europe)):
  x = data[data['Country']==Europe[i]]
  EU_data=pd.concat([x,EU_data])
EU_data
```

```python
# แปลง data  ให้อยู่ในรูปแบบ Transaction data

transactionsEU = EU_data.groupby('InvoiceNo')['StockCode'].apply(list)
```

```python
# หาความถี่ของสินค้าที่ถูกซื้อซ้ำๆกันในทวีปยุโรป โดยมีค่า min support คือ 5%

frequentItemsetsEU = list(apriori(transactionsEU,min_support = 0.05))
frequentItemsetsEU
```

```python
# จำนวนสินค้า
len(frequentItemsetsEU)
```

```python
# หาสินค้าที่ถูกซื้อซ้ำบ่อยในทวีปยุโรป โดยกำหนดค่า min_sup = 0.07  และเช็คว่ามีสินค้ามากกว่า 1 ไอเท็มใน set นั้นหรือไม่

min_sup = 0.03

frequentItem_EU = list(apriori(transactionsEU,min_support = min_sup))
for i in frequentItem_EU:
    if i[0].__len__()>1:
        print('yes')
        print(i)
frequentItem_EU
```
